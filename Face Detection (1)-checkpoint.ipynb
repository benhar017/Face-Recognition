```python
# face detection
face_xml=r"C:\Users\Lenovo\Documents\HCI\face.xml"
```

```python
face_cascade=cv2.CascadeClassifier(face_xml) #calling functions of cascade class.
faces=face_cascade.detectMultiScale(img) #detection of features inside faces
faces
```

```python
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
img2=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```
