```python
from skimage import io
from skimage import color
from skimage import data
#Read image
img = io.imread('C:\\Users\\Lenovo\\Documents\\HCI\\cat.jpg')
#Convert to XYZ
img_xyz = color.rgb2xyz(img)
img_rgb = color.xyz2rgb(img_xyz)
#Show both figures
figure(0)
io.imshow(img_xyz)
figure(1)
io.imshow(img_rgb)
```

```python

#Bezier curve
from skimage import io
from skimage import draw
#Load image
img = io.imread('C:\\Users\\Lenovo\\Documents\\HCI\\cat.jpg')
#Define Bezier curve coordinates
x, y = draw.bezier_curve(0,0, 500, 500, 720,880,100)
#last value is tension
#Draw Bezier curve
img[x, y] = 1
io.imshow(img)
```
